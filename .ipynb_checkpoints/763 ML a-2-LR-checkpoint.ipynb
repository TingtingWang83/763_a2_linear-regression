```python
# One-Hot Encoding
from sklearn.preprocessing import OneHotEncoder
df_processed = pd.get_dummies(df2, prefix_sep="__", columns=['Gender', 'Smoker', 'Location', 'SelfAssessedHealthStatus'])
df_processed
```

```python
df_processed.columns.tolist()
```

```python
#Drop Columns
df_final = df_processed.drop(["Gender__'Male'",'Smoker__0',"Location__'County General Hospital'","SelfAssessedHealthStatus__'Excellent'"],axis=1)
df_final.columns.tolist()
df_final.info
```

```python
# build features
X = df_final[
['Age',
 'Height',
 'Weight',
 "Gender__'Female'",
 'Smoker__1',
 "Location__'St. Mary's Medical Center'",
 "Location__'VA Hospital'",
 "SelfAssessedHealthStatus__'Fair'",
 "SelfAssessedHealthStatus__'Good'",
 "SelfAssessedHealthStatus__'Poor'"]]

y = df_final[['Systolic']]
```

```python
y.info
```

```python
# split train and test data
from sklearn.model_selection import train_test_split

# Divide data into training and validation subsets
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0)
```

```python

# Load dataset......# Create linear regression object
regr= linear_model.LinearRegression()
# Train the model using the training sets
regr.fit(X_train, y_train)
# Make predictions using the testing set
y_pred = regr.predict(X_test)

# The coefficients
print('Coefficients: \n', regr.coef_ )
# The mean squared error
print('MSE: %.2f'  %mean_squared_error(y_test, y_pred))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test, y_pred))
```

```python
# more detail
X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```
