```python
def remove_outlier(df1):
    low = .05
    high = .95
    quant_df1 = df1.quantile([low, high])
    from pandas.api.types import is_numeric_dtype
    for name in list(df1.columns):
        if is_numeric_dtype(df1[name]):
            df1 = df1[(df1[name] > quant_df1.loc[low, name]) & (df1[name] < quant_df1.loc[high, name])]
    return df1
remove_outlier(df1).head()
```
